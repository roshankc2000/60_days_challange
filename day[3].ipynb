# Slicing


You can return a range of characters by using the slice syntax.

Specify the start index and the end index, separated by a colon, to return a part of the string.


```python
#Get the characters from position 2 to 4.

mssg = "Hello, World!"
print(mssg[2:5])
#here 5 wont be included.
```

# Slice From the Start
By leaving out the start index, the range will start at the first character:

```python
mssg = "hello maya"

print (mssg[:5])
```

# Slice To the End
By leaving out the end index, the range will go to the end:


```python
mssg = "Hello Maya"

print (mssg[6:])
```

# Split String
The split() method returns a list where the text between the specified separator becomes the list items.

The split() method splits the string into substrings if it finds instances of the separator:

# STRING FORMAT

As we learned in the Python Variables chapter, we cannot combine strings and numbers like this:

age = 36

txt = "My name is John, I am " + age

print(txt)

<h4>But we can combine strings and numbers by using the format() method!<h3>format()</h3> method takes the passed arguments, formats them, and places them in the string where the placeholders {} are:

```python
age = 2

message = "My name is Roshan , and I am {} years old!"

print (message.format(age))
```

```python
"""
The format() method takes unlimited number of arguments,
and are placed into the respective placeholders:
"""
price = 120
itemno = 73
quantity = 10

order = "I want {} pices of Item {} for {}."

print (order.format(quantity,itemno,price))

```

```python
"""
You can use index numbers {0} to be sure the arguments are 
placed in the correct placeholders:
"""
price = 120
itemno = 73
quantity = 10

order = "I want {1} pices of Item {2} for {0}."

print (order.format(price,quantity,itemno))
```
